### Checking for NULL values in df

```python
# Check for any missing values
df.isnull().sum()
```

### Divide the dataset into Independent(X) and Dependent(Y)variables.

```python
x = df.drop('CLASS', axis = 1)
y = df['CLASS']
```

```python
x.head()
```

### Convert Categorical to Numerical Values if applicable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
x['SHAPE'] = le.fit_transform(x['SHAPE'])
```

```python
x['SOIL_TYPE'] = le.fit_transform(x['SOIL_TYPE'])
```

### Visualizing the Data

```python
# find the ranges of all data
plt.figure(figsize=(30,9))
plt.boxplot(x, labels=x.columns);
```

```python
# find the ranges of all data except proline
x_new = x.drop('PROLINE', axis = 1)
plt.figure(figsize=(16,9))
plt.boxplot(x_new, labels=x_new.columns);
```

```python
# Output variable balance
sns.countplot(x=y);
```

### Split the dataset into training and testing datasets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split( x, y, random_state=0, test_size=0.25, stratify=y)
```

```python
x_train.head()
```

```python
x_test.head()
```

### Applying Decision tree Classification

```python
# import the algorithm of decision tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='gini', random_state=0)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

###

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
```

```python
# Find the accuracy score
accuracy_score(y_test, y_pred)
```

```python
# check the confusion matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# plot the confusion matrix
ConfusionMatrixDisplay(confusion_matrix = cm, display_labels =["High","Low","Medium"] ).plot()
```

```python
print(classification_report(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(16,12))
cn = ["High","Low","Medium"]
plot_tree(model, feature_names=x.columns, filled=True,fontsize=10, rounded=True,class_names=cn);
```
